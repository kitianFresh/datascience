```python
import json
from __future__ import print_function
from bs4 import BeautifulSoup


filename = 'items.txt'
count = 0
news = []
with open(filename) as f:
    for line in f:
        if count > 3:
            break
        item = json.loads(line.decode('utf8'))
        for k, v in item.iteritems():
            print(k, v)
        news.append(item)
        count +=1
        
for item in news:
    item['content'] = BeautifulSoup(item['content']).get_text() if item['content'] else None

for item in news:
    print(item["content"])
    
```
