1.2 解：将(1.1)代入公式(1.4)得：
$$
E(w)=\frac{1}{2}\sum_{n=1}^N\left(\sum_{j=0}^{M}w_jx_n^j-t_n\right)^2+\frac{\lambda}{2}\left\|w\right\|^2\quad(1)
$$
对$w_i$求导,令导函数为0，得：
$$
\sum_{n=1}^N\left(\sum_{j=0}^Mw_jx_n^j-t_n\right)x_n^i+\lambda\sum_{j=0}^Mw_j=0\quad(2)
$$
整理得：
$$
\sum_{j=0}^Mw_j\left(\sum_{n=1}Nx_n^{i+j}+{\lambda}I_{ij}\right)=\sum_{n=1}^Nt_nx_n^i\quad(3)
$$
故$\overline{A_{ij}}=A_{ij}+{\lambda}I_{ij}$,\qquad $T_i=\sum_{n=1}^Nx_n^it_n$.

1.4 证明：首先考虑简单函数$f(x)$,若令$x=g(y)$,则得到一个关于y得函数：
$$
\tilde{f}(y)=f(g(y))\quad(1)
$$
假设$f(x)$在$\hat{x}$处取得最大值，此时$f^\prime(\hat{x})=0$,而函数$\tilde{f}(y)$的最大值在$\hat{y}$处取得，(1)式对y求导，令导函数为0得：
$$
\tilde{f}^\prime(\hat{y})=f^\prime(g(\hat{y}))g^\prime(\hat{y})=0\quad(2)
$$
假设在最大值处$g^\prime(\hat{y})\neq0$，则$f^\prime(g(\hat{y}))=0$。又$f^\prime(\hat{x})=0$,所以取得最大值时变量x和y满足$\hat{x}=g(\hat{y})$。
考虑概率密度函数$p_x(x)$,使用$x=g(y)$做非线性变换，由公式（1.27）得：
$$
p_y(y)=p_x(x)\left|\frac{dx}{dy}\right|=p_x(g(y))\left|g^\prime(y)\right|\quad(3)
$$
令$g^\prime(y)=s\left|g^\prime(y)\right|$,$s\in\{-1,+1\}$,则(3)式可写为：
$$
p_y(y)=p_x(g(y))sg^\prime(y)\quad(4)
$$
(4)式两边对y求导得：
$$
p_y^\prime(y)=sp_x^\prime(g(y))\{g^\prime(y)\}^2+sp_x(g(y))g^{''}(y)\quad(5)
$$
因为(5)式右边第二项的存在，因此$\hat{x}=g(\hat{y})$不再满足，即概率密度的最大值取决于变量的选择。若为线性变换，则由于线性变换的二阶导数为0，即(5)式右边第二项为0，则最大值位置变换方式与变量本身变换方式相同。
